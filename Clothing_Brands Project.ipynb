```python
import requests
from bs4 import BeautifulSoup
```

```python
Titles = []
Prices = []
Brands = []
Fabrics = []
Occasions = []
Colors = []
Country_of_Origins = []
Style_Codes = []
Sizes = []
Return_Time = []
```

```python
base_url = 'https://www.montecarlo.in/collections/men?page='


for page_number in range(1, 5):
    url = base_url + str(page_number)
    web = requests.get(url)
    soup = BeautifulSoup(web.content, 'html.parser')
    

    dress = soup.find_all('div', class_='product-item-meta')
    links = ['https://www.montecarlo.in' + link.find('a')['href'] for link in dress]

  
    for link in links:
        r = requests.get(link)
        soup = BeautifulSoup(r.content, 'html.parser')

   
        title = soup.find('h1', class_='product-meta__title heading h3').text.strip()
        Titles.append(title)

   
        price = soup.find('span', class_='price price--highlight price--large').text.strip()
        Prices.append(price)

        description_container = soup.find('div', class_='yash-blog-f')
        
        


        sizes = soup.find('div', class_='product-form__option-selector').text.strip()
        Sizes.append(sizes)

      
        delivery_container = soup.find('div', class_='product-page-tabs-cts-y')
        delivery = delivery_container.find_all('p')[-1].text.strip()
        Return_Time.append(delivery)

        brand_paragraphs = description_container.find_all('p')
        
        if len(brand_paragraphs) > 0:
            style_code = brand_paragraphs[0].text.strip() if len(brand_paragraphs) > 0 else style_code
            brand = brand_paragraphs[2].text.strip() if len(brand_paragraphs) > 2 else brand
            fabric = brand_paragraphs[3].text.strip() if len(brand_paragraphs) > 3 else fabric
            occasion = brand_paragraphs[6].text.strip() if len(brand_paragraphs) > 6 else occasion
            color = brand_paragraphs[1].text.strip() if len(brand_paragraphs) > 1 else color
            country_of_origin = brand_paragraphs[14].text.strip() if len(brand_paragraphs) > 14 else country_of_origin

        Brands.append(brand)
        Fabrics.append(fabric)
        Occasions.append(occasion)
        Colors.append(color)
        Country_of_Origins.append(country_of_origin)
        Style_Codes.append(style_code)
```

```python


def extract_numeric(price):
    if isinstance(price, str):  
       
        numeric_part = re.search(r'\d[\d.,]*', price)
        if numeric_part:
            # Get the matched numeric part
            numeric_price = numeric_part.group()
            # Replace commas with empty string
            numeric_price = numeric_price.replace(',', '')
            try:
                # Convert the result to a float
                result = float(numeric_price)
                return result
            except ValueError:
                # If conversion fails, print the problematic price
                print(f"Conversion failed for price: {price}")
                # Return None (or any other appropriate value)
                return None
    return price

# Apply the extract_numeric function to the 'Price' column
df['Price'] = df['Price'].apply(extract_numeric)
```

```python
def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False],
                'positive': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(transformed_x, y)
```
